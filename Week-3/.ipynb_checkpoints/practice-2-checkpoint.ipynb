```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread("img/sst.jpeg",0)

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)

rows, cols = img.shape
M = np.float32([[1,0,0],[0,-1,rows],[0,0,1]])
reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))

plt.subplot(1,2,2)
plt.title("Reflected Image")
plt.imshow(reflected_img)

cv.imwrite("reflection_out", reflected_img)
cv.waitKey(0)
cv.destrouAllWindows()
```
