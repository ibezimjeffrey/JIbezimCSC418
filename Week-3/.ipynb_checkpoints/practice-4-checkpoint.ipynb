```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread("img/sst.jpeg",0)
rows, cols = img.shape

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)

cropped_img = img[150:500,200:500]

plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(cropped_img)

cv.imwrite("img/cropped_out.jpeg", cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
