```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


def Laplacian(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    image = cv2.resize(image, (400,400))

    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    laplacian = cv2.convertScaleAbs(laplacian)


    cv2.imshow("Original", image)
    cv2.imshow("laplacian Edges",laplacian)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

def Canny(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    image = cv2.resize(image, (400,400))

    edges = cv2.Canny(image, 100, 200)

    cv2.imshow("original", image)
    cv2.imshow("edges", edges)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

def Robert(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    image = cv2.GaussianBlur(image, (5,5),0)

    kernel_x = np.array([[9,0],[0,-1]])
    kernel_y = np.array([[0,1],[-1,0]])

    robert_x = cv2.filter2D(image,-1,kernel_x)
    robert_y = cv2.filter2D(image, -1, kernel_y)

    robert_combined = np.sqrt(np.square(robert_x) + np.square(robert_y))
    cv2.imshow("Original", image)
    cv2.imshow("Robert Edges", np.uint8(robert_combined))

    cv2.waitKey(0)
    cv2.destroyAllWindows()

def Prewitt(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    image = cv2.resize(image, (500,400))

    kernel_x = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
    kernel_y = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

    prewitt_x = cv2.filter2D(image,-1,kernel_x)
    prewitt_y = cv2.filter2D(image, -1, kernel_y)

    prewitt_combined = np.sqrt(np.square(prewitt_x) + np.square(prewitt_y))
    cv2.imshow("Original", image)
    cv2.imshow("Prewitt Edges", np.uint8(prewitt_combined))

    cv2.waitKey(0)
    cv2.destroyAllWindows()

def Sobel(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    image = cv2.resize(image, (500,400))

    sobel_x = cv2.Sobel(image, cv2.CV_64F,1,0,ksize=3)
    sobel_y = cv2.Sobel(image, cv2.CV_64F,0,1,ksize=3)

    sobel_combined = cv2.magnitude(sobel_x, sobel_y)

    cv2.imshow("Original", image)
    cv2.imshow("Sobel Edges", np.uint8(sobel_combined))

    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
print("Which category of images do you want to see  --Select number : ")
print("1: Contemporary")
print("2: Modern ")
print("3: Traditional ")

category = input("").strip().lower()

match category:
    case "1":
        print("Which image do you want to see : ")
        print("1. Okeosisi Okonkwo's Tree of life")
        print("2. In The Country")
        print("3. Durbar")

        image = input("").strip().lower()
        match image:
            case "1":
                path = "img/contem1.jpg"
            case "2":
                path = "img/contem2.png"
            case "3":
                path = "img/contem3.jpg"



        print("Which edge operator do you want to see on this image --Select number : ")
        print("1: Laplacian operator")
        print("2: Canny operator")
        print("3: Robert operator")
        print("4: Prewitt operator")
        print("5: Sobel operator")
       

        operator = input("").strip().lower()

        match operator:
            case "1":
                Laplacian(path)
            case "2":
                Canny(path)
            case "3":
                Robert(path)
            case "4":
                Prewitt(path)
            case "5":
                Sobel(path)


    case "2":
        print("Which image do you want to see --Select number : ")
        print("1. Untitled 1")
        print("2. Untitled 2")
        print("3. Image IV: panel of miniature")

        image = input("").strip().lower()
        match image:
            case "1":
                path = "img/modern1.jpg"
            case "2":
                path = "img/modern2.jpg"
            case "3":
                path = "img/modern3.jpg"



        print("Which edge operator do you want to see on this image --Select number : ")
        print("1: Laplacian operator")
        print("2: Canny operator")
        print("3: Robert operator")
        print("4: Prewitt operator")
        print("5: Sobel operator")

        operator = input("").strip().lower()

        match operator:
            case "1":
                Laplacian(path)
            case "2":
                Canny(path)
            case "3":
                Robert(path)
            case "4":
                Prewitt(path)
            case "5":
                Sobel(path)

    case "3":
        print("Which image do you want to see  --Select number : ")
        print("1. Gelede Headdress")
        print("2. Unknown")
        print("3. Benin Head")

        image = input("").strip().lower()
        match image:
            case "1":
                path = "img/trad1.jpeg"
            case "2":
                path = "img/trad2.jpg"
            case "3":
                path = "img/trad3.jpg"



        print("Which edge operator do you want to see on this image --Select number : ")
        print("1: Laplacian operator")
        print("2: Canny operator")
        print("3: Robert operator")
        print("4: Prewitt operator")
        print("5: Sobel operator")

        operator = input("").strip().lower()

        match operator:
            case "1":
                Laplacian(path)
            case "2":
                Canny(path)
            case "3":
                Robert(path)
            case "4":
                Prewitt(path)
            case "5":
                Sobel(path)
```
